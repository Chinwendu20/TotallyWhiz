```python
Acc['Hour'] = Acc['Time'].str[0:2]

# convert new column to numeric datetype
Acc['Hour'] = pd.to_numeric(Acc['Hour'])
Acc = Acc.dropna(subset=['Hour'])
# cast to integer values
Acc['Hour'] =Acc['Hour'].astype('int')
```

```python
# define a function that turns the hours into daytime groups
def when_was_it(hour):
    if hour >= 0 and hour < 6:
        return "Early hours of the morning (0-6)"
    elif hour >= 6 and hour < 8:
        return "office and school opening hours (6-8)"
    elif hour >= 8 and hour < 17:
        return "Period for business activities(8-17)"
    elif hour >= 17 and hour < 20:
        return "People start to retire for the day(17-20)"
    else:
        return "night (20-23)"
```

```python
Acc['Daytime'] = Acc['Hour'].apply(when_was_it)
Acc[['Time', 'Hour', 'Daytime']].head(8)
```

```python

sns.set_context("paper")
rc={'font.size': 20, 'axes.labelsize': 20, 'legend.fontsize':20, 
    'axes.titlesize': 30, 'xtick.labelsize': 20, 'ytick.labelsize': 20}
sns.set(style="darkgrid", rc=rc)
sns.catplot(y='Daytime',kind='count',data=Acc,aspect=2,height=10,palette='dark',orient='H')
```

```python
Acc['Day_of_Week'].value_counts()
```

```python
def Day_of_the_week(Position_in_week):
    if Position_in_week==1:
        return "Sunday"
    elif Position_in_week==2:
        return "Monday"
    elif Position_in_week==3:
        return "Tuesday"
    elif Position_in_week==4:
        return "Wednesday"
    elif Position_in_week==5:
        return "Thursday"
    elif Position_in_week==6:
        return "Friday"
    elif Position_in_week==7:
        return "Saturday"
```

```python
Acc['Name_of_day_of_the_week'] = Acc['Day_of_Week'].apply(Day_of_the_week)
Acc[['Name_of_day_of_the_week','Day_of_Week']].head(8)
```

```python
rc={'font.size': 20, 'axes.labelsize': 20, 'legend.fontsize':20, 
    'axes.titlesize': 30, 'xtick.labelsize': 20, 'ytick.labelsize': 20}
sns.set(style="darkgrid", rc=rc)
sns.countplot(x ='Name_of_day_of_the_week'  , data = Acc)
```

```python
# assign the data
fatal   = Acc.Accident_Severity.value_counts()[3]
serious = Acc.Accident_Severity.value_counts()[2]
slight  = Acc.Accident_Severity.value_counts()[1]

names = ['Fatal Accidents','Serious Accidents', 'Slight Accidents']
size  = [fatal, serious, slight]
#explode = (0.2, 0, 0)

# create a pie chart
plt.pie(x=size, labels=names, colors=['red', 'darkorange', 'silver'], 
        autopct='%1.2f%%', pctdistance=0.6, textprops=dict(fontweight='bold'),
        wedgeprops={'linewidth':7, 'edgecolor':'white'})

# create circle for the center of the plot to make the pie look like a donut
my_circle = plt.Circle((0,0), 0.6, color='white')

# plot the donut chart
fig = plt.gcf()
fig.set_size_inches(8,8)
fig.gca().add_artist(my_circle)
plt.title('\nAccident Severity: Share in % (2012-2014)', fontsize=14, fontweight='bold')
plt.show()
```
